```python
# CÃ¡lculo del monto final a pagar con IGV

N = int(input("Ingrese la cantidad de productos: "))

subtotal = 0

for i in range(N):
    precio = float(input(f"Precio del producto {i+1}: "))
    cantidad = int(input(f"Cantidad del producto {i+1}: "))
    subtotal += precio * cantidad

igv = subtotal * 0.18
total = subtotal + igv

print("\n--- RESUMEN DE COMPRA ---")
print(f"Subtotal: S/ {subtotal:.2f}")
print(f"IGV (18%): S/ {igv:.2f}")
print(f"Total a pagar: S/ {total:.2f}")
```
