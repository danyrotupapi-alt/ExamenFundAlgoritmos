```python
while True:
    nombre = input("Ingrese el nombre del estudiante (o escriba FIN para terminar): ")

    if nombre.upper() == "FIN":
        print("Programa finalizado.")
        break

    # Lectura de notas
    n1 = float(input("Ingrese la nota 1: "))
    n2 = float(input("Ingrese la nota 2: "))
    n3 = float(input("Ingrese la nota 3: "))

    # Cálculo del promedio
    promedio = (n1 + n2 + n3) / 3

    # Determinar estado
    if 0 <= promedio <= 10:
        estado = "Desaprobado"
    elif 11 <= promedio <= 14:
        estado = "Recuperación"
    elif 15 <= promedio <= 20:
        estado = "Aprobado"
    else:
        estado = "Promedio no válido"

    # Mensaje final
    print(f"\nHola {nombre}, tu estado es: {estado}\n")
1
```
